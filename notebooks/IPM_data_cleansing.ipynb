```python
# add columns of CRT scores
q1_value = df["q_1_farmers_value"]
q1_score = (q1_value == 3.0).astype("int")
df["q_1_farmers_score"] = q1_score

q2_value = df["q_2_marks_value"]
q2_score = (q2_value == 9.0).astype("int")
df["q_2_marks_score"] = q2_score

q3_value = df["q_3_athletics_value"]
q3_score = (q3_value == 16.0).astype("int")
df["q_3_athletics_score"] = q3_score

df["overall_crt_score"] = df["q_1_farmers_score"] + df["q_2_marks_score"] + df["q_3_athletics_score"]

df[["q_1_farmers_score", "q_2_marks_score", "q_3_athletics_score", "overall_crt_score"]].describe()
```

```python
# assessment result comparing to the ground truth
df["doc_assessment_quality"] = 0
df["doc_ground_truth_value"] = df["doc_ground_truth_value"] - 2
for index, row in df.iterrows():
    if row["doc_ground_truth_value"] == row["doc_overall-truthfulness_value"]:
        df.at[index, "doc_assessment_quality"] = 2
    elif row["doc_ground_truth_value"] * row["doc_overall-truthfulness_value"] > 0:
        df.at[index, "doc_assessment_quality"] = 1
```

```python
# save a subset of relevant features
df_user_profile = df[[
    "unit_id", 
    "q_0_age_answer", 
    "q_0_school_answer", 
    "q_0_taxes_answer", 
    "q_0_political_views_answer", 
    "q_0_consideration_answer", 
    "q_0_southern_border_answer", 
    "q_0_environment_answer", 
    "q_0_age_value",
    "q_0_school_value",
    "q_0_taxes_value",
    "q_0_political_views_value",
    "q_0_consideration_value",
    "q_0_southern_border_value",
    "q_0_environment_value",
    "q_1_farmers_score", 
    "q_2_marks_score", 
    "q_3_athletics_score", 
    "overall_crt_score", 
    "q_0_time_elapsed", 
    "q_1_time_elapsed", 
    "q_2_time_elapsed", 
    "q_3_time_elapsed", 
    "q_0_accesses", 
    "q_1_accesses", 
    "q_2_accesses", 
    "q_3_accesses"
]]
df_user_profile = df_user_profile.drop_duplicates()

df_assessment = df[[
    "unit_id", 
    "doc_ground_truth_value", 
    "doc_overall-truthfulness_value", 
    "doc_confidence_value", 
    "doc_correctness_value", 
    "doc_neutrality_value", 
    "doc_comprehensibility_value", 
    "doc_precision_value", 
    "doc_completeness_value", 
    "doc_speaker's-trustworthiness_value", 
    "doc_informativeness_value",  
    "doc_time_elapsed", 
    "doc_accesses", 
    "doc_assessment_quality"
]]

df_user_profile.to_csv("../data/IPM_worker_profile.csv", index=False)
df_assessment.to_csv("../data/IPM_assessment.csv", index=False)
```
